# Movielens 100k 

### Goal:

I will be attempting to solve the recommendation problem using the Movielens 100k Dataset that I got off of Kaggle. 

The state of the art methods are in the following link: https://paperswithcode.com/sota/collaborative-filtering-on-movielens-100k

The main things I will be researching are the effect of including side data such as user info and movie genre. I will also investigate the effects of the compression of this information as features. 

## Step 1: Import and preprocess data

I will start off with just the ratings data to create the baseline. I will subtract the ratings by 3 so that 3 becomes 0, 1 becomes -2 and so on. The default for missing values will be 0 after the subtraction.

Also, the movieId and userId will be subtracted by 1 just to make the matrix dealings easier so keep that in mind when looking at the MovieId to MovieInformation mapping, for example.  

```python
import pandas as pd
import numpy as np
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'

raw_data = pd.read_csv("./data/Movielens100/u.data", sep = None, names=["userId", "movieId", "rating", "timestamp"])
raw_data = raw_data.loc[:, raw_data.columns != "timestamp"]
#make indices start at 0
raw_data["userId"] -= 1
raw_data["movieId"] -= 1
#make ratings center around 0
raw_data["rating"] -= 3
raw_data
```

```python
# create (943, 1682) matrix of user ratings per movie
user_ratings = pd.DataFrame(np.zeros((943,1682)))
for i in raw_data.index:
    user_ratings[raw_data["movieId"][i]][raw_data["userId"][i]] = raw_data["rating"][i]
user_ratings = user_ratings.to_numpy()    
user_ratings
```

```python
from torch.utils.data import Dataset, DataLoader

class MovielensDataset(Dataset):
    def __init__(self, data):
        self.data = data
        # create a mask of 0 and 1 values where half are 0 and half are 1. 
        #The ratio of masked values is something that can and should be optimized. 
        self.random_mask = np.clip((np.random.randn(1682) + 0.5).round(), a_max = 1, a_min = 0)
        
    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        np.random.shuffle(self.random_mask)
        return self.data[idx] * self.random_mask, self.data[idx]

train_dataset = MovielensDataset(user_ratings[:843])
test_dataset = MovielensDataset(user_ratings[843:])
train_dataloader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=100, shuffle=True)
```

## Build out model : Vanilla Autoencoder

Just while I figure out the proper dimensions for the Autoencoder I plan on using a vanilla Autoencoder to see performance

```python
from torch import nn 
import torch
device = "cuda"
class AutoEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(1682, 1024),
            nn.LeakyReLU(),
            nn.Linear(1024,1024),
            nn.LeakyReLU(),
            nn.Linear(1024,512),
            nn.LeakyReLU(),
            nn.Linear(512,1024),
            nn.LeakyReLU(),
            nn.Linear(1024,1024),
            nn.LeakyReLU(),
            nn.Linear(1024,1682),
            nn.Tanh()
        )
    def forward(self, x):
        ratings = self.linear_relu_stack(x) * 2
        return ratings
    
model = AutoEncoder().to(device)
print(model)
```

```python
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr = 1e-4)
```

```python
losses = []
def train(dataloader, model, loss_fn, optimizer, epoch):
    size = len(dataloader.dataset)
    model.train()
    
    for batch, (X,y) in enumerate(dataloader):
        X, y = X.to(device).to(torch.float32), y.to(device).to(torch.float32)
        
        #compute prediction error 
        pred = model(X)
        loss = loss_fn(pred, y)
        
        #Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if epoch % 100 == 0 and batch % 64 == 0:
            loss , current = loss.item(), (batch+1) * len(X)
            print("Epoch : " + str(epoch))
            print(f"loss: {loss:>7f} [{current:>5d}/{size:>5d}]")
            losses.append(loss)
       
```

```python
test_losses = []
def test(dataloader, model, loss_fn, epoch):
    size = len(dataloader.dataset)
    num_batches= len(dataloader)
    model.eval()
    test_loss, correct = 0,0
    with torch.no_grad():
        for X, y in dataloader:
            X, y = X.to(device).to(torch.float32), y.to(device).to(torch.float32)
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
    test_loss /= num_batches
    if epoch % 100 ==0:
        test_losses.append(test_loss)
        #Come back and actually do the accuracy where it sees if it's at least on the right side of it or compute precision/recall or something like that
        print(f"Test Error: \n Avg Loss : {test_loss:>8f} \n ")
```

```python
epochs = 5000
for t in range(epochs):
    train(train_dataloader, model, loss_fn, optimizer, t)
    test(test_dataloader, model, loss_fn,t )
```

```python
import matplotlib.pyplot as plt 
plt.plot(np.arange(len(losses)), losses)
plt.plot(np.arange(len(losses)), test_losses)
```

## Build out model: Variational Autoencoder

Now we will implement the variational autoencoder and see how it performs

Note: I borrowed some code from https://towardsdatascience.com/variational-autoencoder-demystified-with-pytorch-implementation-3a06bee395ed but modified it a good deal

```python
from torch import nn 
import torch
from sklearn.metrics import precision_recall_fscore_support
device = "cuda"
class VariationalAutoEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Sequential(
            #Encoder
            nn.Linear(1682, 1024),
            nn.LeakyReLU(),
            nn.Linear(1024,1024),
            nn.LeakyReLU(),
        )
        self.decoder = nn.Sequential(
            #Decoder
            nn.Linear(512,1024),
            nn.LeakyReLU(),
            nn.Linear(1024,1682),
            nn.Tanh()
        )
        # distribution parameters
        self.fc_mu = nn.Linear(1024, 512)
        self.fc_var = nn.Linear(1024, 512)

        # for the gaussian likelihood
        self.log_scale = nn.Parameter(torch.Tensor([0.0]))
        
        
    def gaussian_likelihood(self, x_hat, logscale, x):
        scale = torch.exp(logscale)
        mean = x_hat
        dist = torch.distributions.Normal(mean, scale)

        # measure prob of seeing data under p(x|z)
        log_pxz = dist.log_prob(x)
        return log_pxz.sum(dim=1)
    
    def kl_divergence(self, z, mu, std):
        # --------------------------
        # Monte carlo KL divergence
        # --------------------------
        # 1. define the first two probabilities (in this case Normal for both)
        p = torch.distributions.Normal(torch.zeros_like(mu), torch.ones_like(std))
        q = torch.distributions.Normal(mu, std)

        # 2. get the probabilities from the equation
        log_qzx = q.log_prob(z)
        log_pz = p.log_prob(z)

        # kl
        kl = (log_qzx - log_pz)
        kl = kl.sum(-1)
        return kl
    
    def forward(self, x):
        # encode x to get the mu and variance parameters
        x_encoded = self.encoder(x)
        mu, log_var = self.fc_mu(x_encoded), self.fc_var(x_encoded)

        # sample z from q
        std = torch.exp(log_var / 2)
        
        #perform the kernel trick to allow for backprop through sampling
        
        epsilon = torch.distributions.Normal(0, 1).rsample()
        z = mu + epsilon * std
        # decoded
        ratings = self.decoder(z) * 2
        return ratings, z, mu, std
    
    def vae_loss(self, x_hat, x, z, mu, std):
        # reconstruction loss
        recon_loss = self.gaussian_likelihood(x_hat, self.log_scale, x)
    
        # kl
        kl = self.kl_divergence(z, mu, std)

        # elbo
        elbo = (kl - recon_loss)
        elbo = elbo.mean()

        return elbo
    
model = VariationalAutoEncoder().to(device)
print(model)
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr = 1e-4)
mse_loss_fcn = nn.MSELoss()
```

```python
losses = []
mse_losses = []


def get_precision_and_recall(labels, predictions):
    relevance_labels = labels > 0.5
    
    #To evaluate for comparison to other papers we mask by setting 20% of labels to false 
    
    pred_relevance = predictions > 0.5
    
    precision_recall_fscore_support(relevance_labels.flatten().cpu(), pred_relevance.flatten().cpu(), average="binary")
    

def train(dataloader, model,  optimizer, epoch):
    size = len(dataloader.dataset)
    model.train()
    
    for batch, (X,y) in enumerate(dataloader):
        X, y = X.to(device).to(torch.float32), y.to(device).to(torch.float32)
        
        #compute prediction error 
        pred, z, mu, std = model(X)
        loss = model.vae_loss(pred, y, z, mu, std)
        mse_loss = mse_loss_fcn(pred, y)
        
        #Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        if epoch % 100 == 0 and batch % 64 == 0:
            loss , current = loss.item(), (batch+1) * len(X)
            print("Epoch : " + str(epoch))
            print(f"loss: {loss:>7f}")
            print(f"MSE loss: {mse_loss:>7f}")
            losses.append(loss)
            mse_losses.append(mse_loss.item())
            
        
test_losses = []
test_mse_losses = []

def test(dataloader, model, epoch):
    size = len(dataloader.dataset)
    num_batches= len(dataloader)
    model.eval()
    test_loss, correct, test_mse_loss = 0,0,0
    with torch.no_grad():
        for X, y in dataloader:
            X, y = X.to(device).to(torch.float32), y.to(device).to(torch.float32)
            
            
            pred, z, mu, std = model(X)
            test_loss += model.vae_loss(pred, y, z, mu, std).item()
            test_mse_loss += mse_loss_fcn(pred, y)
    test_loss /= num_batches
    test_mse_loss /= num_batches
    if epoch % 100 ==0:
        test_losses.append(test_loss)
        #Come back and actually do the accuracy where it sees if it's at least on the right side of it or compute precision/recall or something like that
        print(f"Test Error: \n Avg Loss : {test_loss:>8f} ")
        print(f" MSE loss: {test_mse_loss:>7f} \n")
        test_mse_losses.append(test_mse_loss.item())
        
        #compute precision and recall
        relevance_labels = y > 0
        pred_relevance = pred > 0.5
        print(precision_recall_fscore_support(relevance_labels.flatten().cpu(), pred_relevance.flatten().cpu(), average="binary"))
```

```python
epochs = 1500
for t in range(epochs):
    train(train_dataloader, model,  optimizer, t)
    test(test_dataloader, model, t )
```
